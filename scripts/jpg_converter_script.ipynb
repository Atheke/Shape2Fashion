```python
from PIL import Image
import os

def convert_images_to_jpg(directory):
    # Iterate over each file in the directory
    for filename in os.listdir(directory):
        # Define full file path
        file_path = os.path.join(directory, filename)

        # Process only jpg and png files
        if filename.lower().endswith('.png') or filename.lower().endswith('.jpg'):
            # Open the image file
            with Image.open(file_path) as img:
                # If it's a PNG, convert and save as JPG
                if filename.lower().endswith('.png'):
                    # Convert the image to RGB (to remove transparency)
                    rgb_img = img.convert('RGB')
                    # Save the image as JPG in the same location
                    new_filename = filename.rsplit('.', 1)[0] + '.jpg'
                    rgb_img.save(os.path.join(directory, new_filename), 'JPEG')
                    print(f"Converted {filename} to {new_filename}")
                else:
                    print(f"{filename} is already a JPG file.")
                    
# Provide the path to the directory containing the images
directory_path = "data2/apple"
convert_images_to_jpg(directory_path)
```

```python
from PIL import Image
import os

def convert_images_to_jpg(directory):
    # Iterate over each file in the directory
    for filename in os.listdir(directory):
        # Define full file path
        file_path = os.path.join(directory, filename)

        # Process only jpg and png files
        if filename.lower().endswith('.png') or filename.lower().endswith('.jpg'):
            try:
                # Attempt to open the image file
                with Image.open(file_path) as img:
                    # If it's a PNG, convert and save as JPG
                    if filename.lower().endswith('.png'):
                        # Convert the image to RGB (to remove transparency)
                        rgb_img = img.convert('RGB')
                        # Save the image as JPG in the same location
                        new_filename = filename.rsplit('.', 1)[0] + '.jpg'
                        rgb_img.save(os.path.join(directory, new_filename), 'JPEG')
                        print(f"Converted {filename} to {new_filename}")
                        # Delete the original PNG file
                        os.remove(file_path)
                        print(f"Deleted original file {filename}")
                    else:
                        print(f"{filename} is already a JPG file.")
            except Exception as e:
                # Print error and skip the file if it cannot be opened
                print(f"Could not open {filename}: {e}")
                
# Provide the path to the directory containing the images
directory_path = "data2/1"
convert_images_to_jpg(directory_path)
```
